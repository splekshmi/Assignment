### 1) Reading the dataset to python environment

```python
import pandas as pd
ds = pd.read_excel('iris.xls')
```

##### Using info() , to get an overall idea about data set 

```python
ds.info()
```

### 2)Displaying the columns in the dataset

```python
print(ds.columns.tolist())      # using tolist() to display the column names as a list 
```

### 3) Calculating the mean of each column of the dataset

```python
 # There are five columns in total,out of which 'classification' is an object/string type.
 # so calculating mean value for the rest 4,setting numeric_only=True 
 # From ds.info(),understood that no column has a nullvalue.
```

```python
round(ds.mean(numeric_only=True),1)   # rounding the result to 1 decimal point similar to other numerical values in the data set
```

### 4) Checking for the null values present in the dataset

```python
if(ds.isna().values.any()==False):                
    print('Null values not present')
else:print('Null values present')  
```

### 5) Visualisations using iris data set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # pyplot is a sublibrary in matplotlib
import seaborn as sns # output of plotting commands displayed inline within Jupyter notebook
```

```python
plt.figure(figsize=(6,4))
plt.title('Distribution of species in the data set',fontsize=18)
sns.countplot(x=ds['Classification'])
plt.xlabel('Species/Classification',fontsize=14) 
plt.ylabel('Count',fontsize=14)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.title('Comparison among various species based on petal length and width',fontsize=18)
sns.scatterplot(x=ds['PL'],y=ds['PW'],hue =ds['Classification'],s= 50)
plt.xlabel('Petal length(PL)',fontsize=12)
plt.ylabel('Petal width(PW)',fontsize=12)


plt.subplot(2,1,2)
plt.title('Comparison among various species based on sepal length and width',fontsize=18)
sns.scatterplot(x=ds['SL'],y=ds['SW'],hue =ds['Classification'],s= 50)
plt.xlabel('Sepal length(SL)',fontsize=12)
plt.ylabel('Sepal width(SW)',fontsize=12)

plt.tight_layout()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('Comparison among species based on various features',loc='left',fontsize=16)
sns.boxplot(x=ds['Classification'],y=ds['SL'])
plt.xlabel('Classification',fontsize=12)
plt.ylabel('Sepal length(SL)',fontsize=12)


plt.subplot(2,2,2)
sns.boxplot(x=ds['Classification'],y=ds['SW'])
plt.xlabel('Classification',fontsize=12)
plt.ylabel('Sepal width(SW)',fontsize=12)



plt.subplot(2,2,3)
sns.boxplot(x=ds['Classification'],y=ds['PL'])
plt.xlabel('Classification',fontsize=12)
plt.ylabel('Petal length(PL))',fontsize=12)



plt.subplot(2,2,4)
sns.boxplot(x=ds['Classification'],y=ds['PW'])
plt.xlabel('Classification',fontsize=12)
plt.ylabel('Petal width(PW)',fontsize=12)


plt.tight_layout()
```

```python
plt.figure(figsize=(8,6))
plt.title('Correlation among petal length , petal width,Sepal length and sepal width',fontsize=12)
sns.heatmap(ds.corr(),annot=True)  # annot used to show annotations inside the matrix
```
